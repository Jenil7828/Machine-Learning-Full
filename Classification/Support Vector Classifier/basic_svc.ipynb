# Support Vector Machine classifier

```python
import pandas as pd
import numpy as numpy
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Lets create a sample DataFrame
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=1, n_redundant=0)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
Xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
svc = SVC(kernel='linear')
svc.fit(Xtrain, ytrain)
```

```python
y_pred = svc.predict(xtest)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(ytest, y_pred))
```

```python
print(confusion_matrix(ytest, y_pred))
```

# Radial Basis Function vs Linear for non-linear data

```python
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2, n_redundant=0)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
Xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.25, random_state=10)
rbf=SVC(kernel='rbf')
rbf.fit(Xtrain, ytrain)
```

```python
y_pred = rbf.predict(xtest)
y_pred_l = svc.predict(xtest)
```

```python
print(classification_report(ytest, y_pred))
```

```python
print(classification_report(ytest, y_pred_l))
```

# Polynomial vs Linear Kernel Function

```python
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2, n_redundant=0)
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
Xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.25, random_state=10)
poly = SVC(kernel='poly')
poly.fit(Xtrain, ytrain)
```

```python
y_pred_poly = poly.predict(xtest)
y_pred_lin = svc.predict(xtest)
```

```python
print(classification_report(ytest, y_pred_poly))
print(classification_report(ytest, y_pred_lin))
```

# Other Parameters and Grid Search CV

```python
from sklearn.model_selection import GridSearchCV

# defining parameter range
param_grid = {
    'C':[0.1, 1, 10, 100, 1000],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['rbf']
    }
```

```python
grid = GridSearchCV(SVC(), param_grid=param_grid, refit=True, cv=5, verbose=3)
```

```python
grid.fit(Xtrain, ytrain)
```

```python
grid.best_params_
```

```python
y_pred_poly = poly.predict(xtest)
y_pred_lin = svc.predict(xtest)
y_pred_rbf = rbf.predict(xtest)
y_pred = grid.predict(xtest)
print(classification_report(ytest, y_pred_poly))
print(classification_report(ytest, y_pred_lin))
print(classification_report(ytest, y_pred_rbf))
print(classification_report(ytest, y_pred))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print("Predictions:", y_pred)
```

```python
model.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(y_pred, bins=30, alpha=0.7, color='blue', label='Predicted')
plt.hist(y_test, bins=30, alpha=0.7, color='orange', label='Actual')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.title('Distribution of Predicted vs Actual Classes')
plt.legend()
plt.show()
```

```python


plt.figure(figsize=(8, 5))
plt.plot(X_test, y_test, label='Sigmoid Function')
plt.title('Sigmoid Function Visualization')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Visualize sigmoid function for X_test and y_test
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Use the first feature of X_test for visualization
x_feature = X_test[:, 0]
y_sigmoid = sigmoid(x_feature)

plt.figure(figsize=(8, 5))
plt.scatter(x_feature, y_test, color='orange', label='Actual y_test', alpha=0.6)
plt.scatter(x_feature, y_sigmoid, color='blue', label='Sigmoid(X_test[:,0])', alpha=0.6)
plt.title('Sigmoid Function on X_test Feature vs Actual y_test')
plt.xlabel('X_test (first feature)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
